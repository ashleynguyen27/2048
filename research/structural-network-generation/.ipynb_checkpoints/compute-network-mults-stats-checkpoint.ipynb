```python
import json
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
mults_folder = "/Users/amitvakula/Documents/code/ucsf/structural-network-generation/network-multipliticies"
save_folder = os.path.join(mults_folder, "computed-summaries")
example_json = os.path.join(mults_folder, "204420-network-multiplicities.json")
j_dict = json.load(open(example_json, 'r'))
```

```python
def parse_json_network(json_path):
    loaded_dict = json.load(open(json_path, 'r'))
    to_rtn = {}
    for key in loaded_dict:
        locals = {}
        exec("t = {0}".format(key), {}, locals)
        to_rtn[locals['t']] = loaded_dict[key]
    return to_rtn

def dump_json_network(network, json_path):
    writeable_network = {}
    for edge in network:
        writeable_network[str(edge)] = network[edge]
    f = open(json_path, 'w')   
    json.dump(writeable_network, f)
    f.close()

example_parsed_dict = parse_json_network(example_json)
```

```python
subjid_to_network = {}
it = glob.glob(os.path.join(mults_folder,"*.json"))
for json_path in it:
    parsed_dict = parse_json_network(json_path)
    subjid = int(json_path.split(os.sep)[-1].split("-")[0])
    subjid_to_network[subjid] = parsed_dict
```

```python
mult_distributions = {}
for subjid in subjid_to_network:
    network = subjid_to_network[subjid]
    for roi_pair in network:
        if not (roi_pair in mult_distributions):
            mult_distributions[roi_pair] = []
        mult_distributions[roi_pair].append(network[roi_pair])
```

```python
distr_file = os.path.join(save_folder, "multiplicity-distributions.json")
f = open(distr_file, 'w')
dump_json_network(mult_distributions, distr_file)
f.close()

print(mult_distributions)
```

```python
# compute averages.json
roipair_to_avgs = {}
roipair_to_mults = {}
for roi_pair in mult_distributions:
    vals = mult_distributions[roi_pair]
    roipair_to_avgs[roi_pair] = np.average(vals)
    roipair_to_mults[roi_pair] = len(vals)
```

```python
avg_mult_file = os.path.join(save_folder, "network-average-multiplicities.json")
f = open(avg_mult_file, 'w')
dump_json_network(roipair_to_avgs, avg_mult_file)
f.close()

print(roipair_to_avgs)
```

```python
mult_samplesize_file = os.path.join(save_folder, "network-edge-sample-size.json")
f = open(mult_samplesize_file, 'w')
dump_json_network(roipair_to_mults, mult_samplesize_file)
f.close()

print(roipair_to_mults)
```

```python
for roi_pair in roipair_to_mults:
    mult = roipair_to_mults[roi_pair]
    avg = roipair_to_avgs[roi_pair]
    plt.scatter(avg, mult)
plt.xlabel("average number of streamlines")
plt.ylabel("number of networks which have data for the roi pair")
plt.show()
```
